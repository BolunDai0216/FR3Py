```python
from FR3Py.lcm_msgs.fr3_commands import fr3_cmd
from FR3Py.lcm_msgs.fr3_states import fr3_state
from FR3Py.sim.lcm_bridges import LCMBridgeClient
from FR3Py.utils import NumpyMemMapDataPipe
import numpy as np
```

```python
import yaml

def load_config(file_path):
    with open(file_path, 'r') as file:
        config = yaml.safe_load(file)
    return config
```

```python
cfg = load_config("/home/mocap/manipulation/rooholla/FR3Py/FR3Py/sim/isaac/sim_config.yaml")
```

```python
camera_pipes = []
for camera in cfg['cameras']:
    if camera['type'] == 'rgb':
        shape = (camera['resolution'][1], camera['resolution'][0], 4)
    else:
        shape = (camera['resolution'][1], camera['resolution'][0])
    camera_pipes.append(NumpyMemMapDataPipe(camera['name']+'_'+camera['type'], force = False, shape= shape))
    print(camera['name']+'_'+camera['type'])
```

```python
lcm_bridge = LCMBridgeClient(robot_name='fr3')
```

```python
import cv2
cmd  = fr3_cmd()

while True:
    state = lcm_bridge.getStates(timeout=0.1)
    cmd.cmd =  9*[np.random.randn(1)*1]
    lcm_bridge.sendCommands(cmd)
    imgs =  [pipe.read() for pipe in camera_pipes]
    for i, img in enumerate(imgs):
        cv2.imshow(f'{i}', img)
    cv2.waitKey(1)
```

```python
# load the simulation configs
cfg = load_config(
    "/home/mocap/manipulation/rooholla/FR3Py/FR3Py/sim/isaac/sim_config.yaml"
)
robots = cfg["robots"]
cameras = cfg["cameras"]
env_cfg = cfg["environment"]
```

```python
world = World(
    physics_dt=cfg["environment"]["simulation_dt"],
    rendering_dt=cfg["environment"]["rendering_dt"],
)

# load the scenerendering_dt
assets_root_path = get_assets_root_path()
if assets_root_path is None:
    print("Could not find Isaac Sim assets folder")
```

```python
# load the scenerendering_dt
assets_root_path = get_assets_root_path()
if assets_root_path is None:
    print("Could not find Isaac Sim assets folder")

prim = get_prim_at_path(env_cfg["prim_path"])
if not prim.IsValid():
    prim = define_prim(env_cfg["prim_path"], "Xform")
    asset_path = (
        assets_root_path + env_cfg["usd_path"]
        if env_cfg["buildin"] is True
        else env_cfg["usd_path"]
    )
    prim.GetReferences().AddReference(asset_path)

# spawn table
if "tables" in cfg.keys():
    tables = cfg["tables"]
    prim = get_prim_at_path(tables[0]["prim_path"])
    if not prim.IsValid():
        prim = define_prim(tables[0]["prim_path"], "Xform")
        asset_path = (
            assets_root_path + tables[0]["usd_path"]
            if tables[0]["buildin"] is True
            else tables[0]["usd_path"]
        )
        prim.GetReferences().AddReference(asset_path)
        UsdGeom.Xformable(prim).AddTranslateOp().Set(
            (
                tables[0]["position"][0],
                tables[0]["position"][1],
                tables[0]["position"][2],
            )
        )

        UsdGeom.Xformable(prim).AddRotateXYZOp().Set(
            (
                tables[0]["orientation"][0],
                tables[0]["orientation"][1],
                tables[0]["orientation"][2],
            )
        )
```

```python
# spawn a FR3 robot
fr3 = world.scene.add(
    FR3(
        prim_path=robots[0]["prim_path"],
        usd_path=robots[0]["usd_path"],
        name=robots[0]["name"],
        position=np.array(robots[0]["position"]),
        physics_dt=cfg["environment"]["simulation_dt"],
    )
)
world.reset()
fr3.initialize()
```
